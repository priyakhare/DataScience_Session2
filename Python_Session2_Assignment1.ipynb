```python
#***************************************************************************************
#
#   Program2  :  Creating a Pattern with Nested For Loop
#
#***************************************************************************************


# Initializing the variables for pattern and counter

symStar = "*"

cnt = 0

# Nested For Loop for creating the pattern , Outer For Loop is to print the number of rows 
# and the inner loop for printing the Asterics  in the each row 

for i in range(0,9):
    if i < 4: 
        ntimes = i+1
    elif i >= 4:
        ntimes = 5 - cnt
        cnt += 1
        
    for j in range(0,ntimes):
        print(symStar , end= " ")
        j += 1
    i += 1
    print("\n")
    
 #***************************************************************************************
```

```python
#*****************************************************************************************
#  Program 3  : Write a Python program to reverse a word after accepting the input from the user. 
#
#******************************************************************************************

# Taking the input from User for reversing the string entered 
inpVal  = input("Please enter a word to be reversed : ")

# Using String Slicing function to reverse the string
revVal =  inpVal[::-1]

print("The reversed word is : " + revVal )












#******************************************************************************************
```
